```python
def roi(revenue, cost):
    assert isinstance(revenue, (int, float)), "Revenue must be numeric"
    assert isinstance(cost, (int, float)), "Cost must be numeric"
    assert cost > 0, "Cost must be greater than 0"

    return (revenue - cost) / cost
```

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float))
  assert isinstance(annual_rate, (int, float))
  assert isinstance(months, int)

  if principal <= 0 or annual_rate < 0 or months <= 0:
    raise ValueError("Invalid input values")

  try:
    r = annual_rate / 12
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1+r) ** (-months))
  except Exception as e:
    raise Exception(f"Error: {e}")


  return pmt
```

```python
def getValues(n):
    values = []
    i = 0

    while i < n:
        try:
            num = float(input(f"Enter value {i+1}: "))

            if num < 0:
                raise ValueError("Value cannot be negative")

            values.append(num)
            i += 1

        except ValueError:
            print("Invalid input. Please enter a non-negative number.")

    return values


def calcAverage(alist):
    # Exclude first 3 values
    trimmed = alist[3:]

    if len(trimmed) == 0:
        return "Not enough values to calculate average"

    avg = sum(trimmed) / len(trimmed)
    return avg


# Main program
n = int(input("How many values? "))

data = getValues(n)
result = calcAverage(data)

print("\nAverage (excluding first 3 values):", result)
```
