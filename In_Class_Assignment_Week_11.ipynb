```python
def roi(revenue, cost):
  assert cost != 0
  assert isinstance(revenue, (int, float))
  assert isinstance(cost, (int, float))
  return (revenue - cost) / cost
```

```python
def loan_payment(principal, annual_rate, months): #part1 add assertions
  assert isinstance(principal, (int, float))
  assert isinstance(annual_rate, (int, float))
  assert isinstance(months, int)

  if principal <= 0 or annual_rate < 0 or months <= 0:
    raise ValueError("Invalid input values")

  try:
    r = annual_rate / 12
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1+r) ** (-months))
  except Exception as e:
    raise Exception(f"Error: {e}")


  return pmt
```
