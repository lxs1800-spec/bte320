```python
sales = float(input ("Enter the sale amount ($): "))
rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gross_profit= sales - cogs
taxes = rate/100 * gross_profit
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods sold: {cogs}$")
print(f"Gross profit: {gross_profit}$")
print(f"Taxes:{taxes}$")
print(f"Net income: {net}$")
```

```python
import math


D = float(input("Enter annual demand (units/year): "))
S = float(input("Enter ordering cost ($/order): "))
H = float(input("Enter holding cost ($/year/unit): "))


eoq = math.sqrt((2 * D * S) / H)

print(f"\nEconomic Order Quantity (EOQ): {eoq:,.2f} units")
```
