```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Choose numerical operator: ")

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    result = num1 / num2
elif operator == "//":
    result= num1// num2
elif operator == "**":
    result = num1 ** num2
elif operator == "%":
    result = num1 % num2
else:
  print("Invalid operation")
  result= None

if result is not None:
  print(f"Result: {result}")
```
