```python
class Greeting:
  def __init__(self, first, last): #first parameter is always self
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(f"{self.first}")

  def __str__(self):
    return "Class Greeting\nReturns formal or informal greetings"

s = Greeting("John", "Doe")
print(s)
s.display()
s.display(formal=True)
```
