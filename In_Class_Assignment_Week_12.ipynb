```python
class Greeting:
  def __init__(self, first, last): #first parameter is always self
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(f"{self.first}")

  def __str__(self):
    return "Class Greeting\nReturns formal or informal greetings"

s = Greeting("John", "Doe")
print(s)
s.display()
s.display(formal=True)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self._courses = []

  def get_courses(self):
    return self._courses

  def get_name(self):
    return self._name

  def set_name(self, newName):
    self._name = newName

  def _len_(self):
    return len(self._courses)

  def enroll(self, course):
    if course in self._courses
      print(f"Student {self._name} has already enrolled in course {course}.")
    else:
      self._courses.append(course)
```

```python
s = Student("Jane Doe", 12345)
print("Student name", s.get_name())
s.get_name("Jane B. Doe")
print("Student name:", s.get_name())
print("Courses enrolled in:", s.get_courses())
s.enroll("BTE320")
s.enroll("BTE320")
print(f"Student {s.get_name()} has enrolled in {len(s)} courses: {s.get_courses()}")
```
