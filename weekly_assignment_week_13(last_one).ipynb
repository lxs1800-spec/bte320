```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):  #use other bc theyre in the same class
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5


class Rectangle:
    def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4):
        self.p1 = Point(x1, y1)
        self.p2 = Point(x2, y2)
        self.p3 = Point(x3, y3)
        self.p4 = Point(x4, y4)

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height
```

```python
square = Rectangle(0, 3, 3, 3, 0, 0, 3, 0)
square.area()
```
