```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #return s[-1] + reverse(s[:-1]) #take the last char and put it first OR
    #return reverse(s[1:]) + s[0] #take the first char and put it last  OR
    return s[-1] + reverse(s[1:-1]) + s[0]

reverse('12345')
```
