```python
numbers = []

while True:
  num = float(input("Give a number: "))
  if num == 0:
    break
  numbers.append(num)
numbers.sort(reverse=True)
print(numbers)
```

```python
weights = [10.0, 9.9, 10.1, 10.05]

print(sum(weights) / len(weights)) # avg weight
print(max(weights)) # largest value
```

```python
products = {
  "P-01": 9.95,
  "P-02": 10.5
}

id = input ("Enter product ID: ")

if id in products:
  print(products[id])
else:
  print(f"Product {id} not found...")
```

```python
names = {}

while True:
  name = input("Give a name: ")
  if name == "":
    break

  if name in names:
    names[name] = names[name] + 1
  else:
    names[name] = 1

print(names)
```
