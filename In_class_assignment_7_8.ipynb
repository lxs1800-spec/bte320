```python
N = 5

counts = {}

i = 0
while i < N:
    name = input("Enter a name: ")

    if name not in counts:
        counts[name] = 1
    else:
        counts[name] += 1

    i += 1

print("\nFinal dictionary:")
print(counts)
```

```python
products = {
    "P-01": 9.95,
    "P-02": 10.50
}


product_id = input("Enter product ID: ")


if product_id in products:
    print(product_id, products[product_id])
else:
    print(product_id, "not found")
```
